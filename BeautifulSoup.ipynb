```python
#import library yang dibutuhkan
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#buatlah request ke website
website_url = requests.get('https://id.wikipedia.org/wiki/Demografi_Indonesia').text
soup = BeautifulSoup(website_url, 'html.parser')
```

```python
#ambil table dengan class 'wikitable sortable'
my_table = soup.find('table', {'class':'wikitable sortable'})
```

```python
table = pd.read_html(str(my_table))
```

```python
my_table = pd.DataFrame(table[0])
```

```python
my_table
```

```python
my_table = my_table.drop((my_table.tail(1)).index,axis=0)
```

```python
my_table = my_table.rename(columns={'Luas (km²)[11]':'Luas (km²)','Populasi (2010)[12]':'Populasi (2010)',
                                    'Populasi (2020)[9]':'Populasi (2020)'})
```

```python
my_table
```

```python
my_table['Luas (km²)'] = my_table['Luas (km²)'].str.replace('.','',regex=False).str.replace(',','.',regex=False)
```

```python
my_table['Populasi (2010)'] = my_table['Populasi (2010)'].str.replace('.','',regex=False).astype(int)
```

```python
my_table['Populasi (2020)'] = my_table['Populasi (2020)'].str.replace('.','',regex=False).astype(int)
```

```python
my_table
```

```python
my_table['Luas (km²)'] = my_table['Luas (km²)'].astype(float)
```

##### Kepadatan Penduduk menunjukkan banyaknya penduduk yang mendiami setiap km²

```python
my_table['Kepadatan Penduduk'] = my_table['Populasi (2020)'] / my_table['Luas (km²)']
```

```python
my_table
```

```python
Luas_wilayah_Indonesia = my_table['Luas (km²)'].sum()
Luas_wilayah_Indonesia
```

#### 5 Provinsi dengan Tingkat Kepadatan Penduduk Rendah

```python
my_table['Pertambahan Penduduk'] =  (my_table['Populasi (2020)'] - my_table['Populasi (2010)'])
```

```python
my_table
```

#### 5 Provinsi dengan Pertambahan Penduduk Tertinggi (2020-2010)

```python
my_table.sort_values(by='Pertambahan Penduduk',ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
Sumatera = my_table['Nama provinsi'].iloc[0:10].copy().to_list()
```

```python
Jawa = my_table['Nama provinsi'].iloc[10:16].copy().to_list()
```

```python
Bali_NusaTenggara = my_table['Nama provinsi'].iloc[16:19].copy().to_list()
```

```python
Kalimantan = my_table['Nama provinsi'].iloc[19:24].copy().to_list()
```

```python
Sulawesi = my_table['Nama provinsi'].iloc[24:30].copy().to_list()
```

```python
Maluku = my_table['Nama provinsi'].iloc[30:32].copy().to_list()
```

```python
Papua = my_table['Nama provinsi'].iloc[32:].copy().to_list()
```

```python
Papua
```

```python
Area = []
for i in my_table['Nama provinsi']:
    if i in(Sumatera):
        Area.append('Sumatera')
    if i in(Jawa):
        Area.append('Jawa')
    if i in(Bali_NusaTenggara):
        Area.append('Bali_NusaTenggara')
    if i in(Kalimantan):
        Area.append('Kalimantan')
    if i in(Sulawesi):
        Area.append('Sulawesi')
    if i in(Maluku):
        Area.append('Maluku')
    if i in(Papua):
        Area.append('Papua')    
```

```python
len(Area)
```

```python
my_table.insert(6,'Area',Area)
```

```python
my_table
```

```python
my_table.to_csv("Demografi_Indonesia.csv"
```

```python
plt.figure(figsize=(15,10))
my_table.groupby('Area')['Kepadatan Penduduk'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Grafik Kepadatan Penduduk berdasarkan Pulau', fontsize=12)
plt.ylabel('Tingkat Kepadatan')
plt.xticks(rotation=0)
plt.grid(linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(15,10))
my_table.groupby('Area')['Luas (km²)'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Grafik Luas Wilayah berdasarkan Pulau', fontsize=12)
plt.ylabel('Luas (km²)')
plt.xticks(rotation=0)
plt.show()
```
